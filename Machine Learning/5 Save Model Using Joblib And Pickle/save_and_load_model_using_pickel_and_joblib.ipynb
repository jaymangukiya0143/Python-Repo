```python
# Pickle and Joblib works in a similar way but Joblib is more efficient when having big data.
# It doesn't save numpy arrays or dataframes for data. It saves m's and b's generated by training datasets boz that is enough to predict.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model

import pickle
import joblib
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict(np.array(3300).reshape(1,1))
```

```python
print(reg.coef_)
print(reg.intercept_)
print('Predicted price is {}'.format(reg.predict(np.array([5000]).reshape(1,1))))
```

```python
# PICKLE
```

```python
# Saving

with open('model_pickle','wb') as f: #wb is write binary
    pickle.dump(reg, f)
```

```python
# Opening saved model

with open('model_pickle','rb') as f: #rb is read binary
    saved_model = pickle.load(f)
    
saved_model.predict(np.array([4000]).reshape(1,1))
```

```python
# JOBLIB
```

```python
# Saving 

joblib.dump(reg,'model_joblib')
```

```python
# Loading

saved_joblib_model = joblib.load('model_joblib')
saved_joblib_model.predict(np.array([3500]).reshape(1,1))
```
