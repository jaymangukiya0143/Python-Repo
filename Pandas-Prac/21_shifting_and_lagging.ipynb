```python
# Shifting and Lagging

import pandas as pd
```

```python
df = pd.read_csv('F:/Jay/Python/Code Basics Youtube/Pandas/21_fb.csv',index_col='Date', parse_dates=True)
df
```

```python
%matplotlib inline

df.plot()
```

```python
df.shift(1).plot()
```

```python
df.shift(-2).plot()
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['1 Day Price Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 Day Percentage Return'] = (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
df = df[['Price']] # will remove all columns except 'Price'
print(df)
print(df.index)
df.index = pd.date_range(start='2017-08-15', periods=10, freq='B')
print(df.index)

df.tshift(1) # to shift dates
```
