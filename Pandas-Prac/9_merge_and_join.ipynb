```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'city':['New York','Chicago','Orlando','Baltimore'],
    'temperature':[21,14,35,32]
})
df1
```

```python
df2 = pd.DataFrame({
    'city':['New York','Chicago','San Francisco'],
    'humidity':[65,68,75]
})
df2
```

```python
df3 = pd.merge(df1,df2, on='city') # same as inner join (intersection)
df3
```

```python
df3 = pd.merge(df1,df2,on='city', how='outer') #outer join (union)
df3
```

```python
df3 = pd.merge(df1,df2,on='city', how='right', indicator=True) #right inner join
df3
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2, on='city')
df3
```

```python
df3 = pd.merge(df1,df2, on='city', suffixes=('_first','_second'))
df3
```

```python
# JOIN

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2, lsuffix='_l', rsuffix='_r')
```
