```python
# Pandas Time Series Part-2 (date_range)

import pandas as pd
```

```python
df = pd.read_csv('F:/Jay/Python/Code Basics Youtube/Pandas/16_aapl_no_dates.csv')
df.head()
```

```python
rng = pd.date_range(start='6/1/2017',end='6/30/2017',freq='B')
rng
```

```python
df.set_index(rng, inplace=True)
df.head()
```

```python
%matplotlib inline
df['2017-06-01':'2017-06-10'].Close.plot()
```

```python
df.asfreq('H', method='pad')
```

```python
df.asfreq('D', method='pad')
```
