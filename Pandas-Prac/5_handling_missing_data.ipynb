```python
import pandas as pd
```

```python
df = pd.read_csv("F:/Jay/Python/Code Basics Youtube/Pandas/weather_data2.csv", parse_dates=['day'])
df.set_index('day', inplace=True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df = df.fillna(method='ffill') #ffill = forward fill
new_df
```

```python
new_df = df.fillna(method='bfill') #ffill = forward fill
new_df
```

```python
new_df = df.fillna(method='ffill', limit=1)
new_df
```

```python
new_df = df.fillna(method='bfill', axis='columns') 
new_df
```

```python
new_df = df.interpolate() #method is 'linear' by  default
new_df 
```

```python
new_df = df.interpolate(method='time')
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1) #thresh = 1 means rows with atleast one valid value
new_df
```

```python
date_range = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(date_range)
new_df = df.reindex(idx)
new_df
```
