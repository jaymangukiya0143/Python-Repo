```python
# Read from excel files

df = pd.read_excel('F:/Jay/Python/Code Basics Youtube/Pandas/stock_data.xlsx', 'Sheet1')
df
```

```python
def convert_people_cell(cell):
    if cell=="n.a.":
        return 'Wal Carlton'
    else:
        return cell
    
def convert_price_cell(cell):
    if cell=="n.a.":
        return 1000
    else:
        return cell
    
df = pd.read_excel('F:/Jay/Python/Code Basics Youtube/Pandas/stock_data.xlsx', 'Sheet1', converters={
    'people': convert_people_cell,
    'price': convert_price_cell
})
df
```

```python
# Writing excel file

df.to_excel('F:/Jay/Python/Code Basics Youtube/Pandas/written_stock_data.xlsx', sheet_name='Stocks')
```

```python
# Write two dataframes in an excel file in two diff sheets 

df_stocks = pd.DataFrame({
    'tickers':['GOOGLE','WMT','MICROSOFT'],
    'price':[1000,1200,1500]
})
df_weather = pd.DataFrame({
    'day':['1/1/2017','1/2/2017','1/3/2017'],
    'temperature':[35,36,37]
})

with pd.ExcelWriter('F:/Jay/Python/Code Basics Youtube/Pandas/written_excel_file_two_sheets.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name='Stocks')
    df_weather.to_excel(writer, sheet_name="Weather")
```
